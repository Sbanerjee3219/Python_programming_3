#### Write a Python Program to Check if a Number is Positive, Negative or Zero?

```python
n=int(input("enter number of your choice "))
if n>0:
    print("number is greater than zero and the number is: ",n)
elif n<0:
    print("number is lesser than zero and the number is: ",n)
else:
    print("the number is zero")
```

#### Write a Python Program to Check Leap Year?

```python
year= int(input("enter year: "))
if year%4==0 and year%100==0 and year%400==0:
    print("the year {} is leap year".format(year))
elif year%4==0 and year%100!=0:
    print("the year {} is leap year".format(year))

else:
    print("the year is not leap year")
            
```

#### Write a Python Program to Check Prime Number?


```python
n=int(input("enter number"))
i=2
while i<int(n/2):
    if n%i==0:
        print("not prime")
        break
    i+=1    
else:
    print("it's a prime number")
```

#### Write a Python Program to Print all Prime Numbers in an Interval of 1-10000?

```python
for i in range(1, 10001):
    for j in range(2, int(i/ 2)+2):
        if i % j == 0:
            break
    else:
        print("The prime numbers are: ", i)
```
